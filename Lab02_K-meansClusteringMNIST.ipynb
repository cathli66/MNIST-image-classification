```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)
# print(mnist.keys())
X, y = mnist["data"], mnist["target"]
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
k = 45

# print(y_train[0])
# x_image = X_train[0].reshape(28, 28)
# plt.imshow(x_image, cmap="binary")
# plt.axis("off")
# plt.show()

kmeans = KMeans(n_clusters=k)
kmeans.fit(X_train)

clusterToDigit = []
for i in range(k):
    freqs = [0]*k
    for idx, label in enumerate(kmeans.labels_):
        if label == i:
            freqs[int(y_train[idx])] += 1
    digitLabel = freqs.index(max(freqs))
    clusterToDigit.append(digitLabel)
    print("Cluster", i, "contains the number", digitLabel)

# numCorrect = sum([1 for idx, test in enumerate(X_test) if clusterToDigit[kmeans.predict([test])[0]] == int(y_test[idx])])
numCorrect = 0
total = len(X_test)
for idx, test in enumerate(X_test):
    if clusterToDigit[kmeans.predict([test])[0]] == int(y_test[idx]):
        numCorrect += 1
print("k =", k)
print("Accuracy:", numCorrect/total)
```
